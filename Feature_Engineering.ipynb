```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
features = [feature for feature in df.columns if df[feature].dtype=='O']
features
```

```python
category = df[features]
category.head()
```

```python
for col in category.columns:
    print(f"{col} : {len(category[col].unique())} labels.")
```

```python
pd.get_dummies(category,drop_first=True).shape
```

Note: sometimes we have many categories in a feature which may create that many columns and at the can reduce the performance of the model so we select some categories(let's say 10) and for remaining other we make a new category.

```python
category.X2.value_counts().sort_values(ascending=False).head(10)
```

```python
top_10 = [x for x in category.X2.value_counts().sort_values(ascending=False).head(10).index]
top_10
```

```python
data = pd.DataFrame(category['X2'],index='X2')
data
```

```python
for label in top_10:
    category[label] = np.where(category['X2']==label,1,0)
category[['X2']+ top_10].head()
```

```python
def one_hot_encoding(df):
    for col in df.columns:
        top_10 = [x for x in category[col].value_counts().sort_values(ascending=False).head(10).index ]
        for label in top_10:
            df[col+"_"+label] = np.where(df[col]==label,1,0)
    return df
```

```python
data = one_hot_encoding(category)
data.head()
```
